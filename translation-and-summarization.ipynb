```python
from transformers.utils import logging
from transformers import pipeline 
import torch
```

```python
translator = pipeline(task="translation",
                      model="facebook/nllb-200-distilled-600M",
                      torch_dtype=torch.bfloat16) 
```

```python
text = """\
My parrot is colorful, \
Your bunny is gentle.
Her turtle is wise.
His hamster is curious. \
We all cherish our lovely pets!"""
```

```python
text_translated = translator(text,
                             src_lang="eng_Latn",
                             tgt_lang="tur_Latn")
```

```python
text_translated
```

```python
summarizer = pipeline(task="summarization",
                      model="facebook/bart-large-cnn",
                      torch_dtype=torch.bfloat16)
```

```python
text = """Istanbul is the most populous city in Turkey, renowned as the historical, economic, 
and cultural heart of the country. It has an estimated population of over 17 million 
residents, spread across an area of about 5,343 square kilometers (2,063 square miles). Straddling the Bosporus Strait, 
Istanbul bridges Asia and Europe both physically and culturally, making it a city of significant geopolitical importance. 
The city's vast area encompasses a rich diversity of neighborhoods, historical sites, and cultural attractions, 
from the ancient walls of Constantinople to the modern skyline marked by skyscrapers.

Istanbul serves as the main cultural and financial center of Turkey and encompasses 
the entire Istanbul Province, which is home to approximately 16 percent of Turkey's population. 
The city's unique position has made it an epicenter for trade and cultural exchange for centuries, contributing to its vibrant and 
diverse atmosphere. 
Istanbul's economy is robust, with a dynamic mix of commerce, arts, technology, and tourism. The city hosts the headquarters of 
numerous Turkish companies and multinational corporations.

Historically known as Byzantium and later Constantinople, Istanbul has been at the crossroads of civilizations. Its rich history is 
reflected in its majestic architecture, including famous landmarks such as the Hagia Sophia, Topkapi Palace, and the Blue Mosque. 
These historical monuments, along with the bustling Grand Bazaar and the scenic Bosphorus, attract millions of tourists each year, 
making tourism a significant part of the city's economy."""
```

```python
summary = summarizer(text,
                     min_length=10,
                     max_length=100)
```
